```python
# using geopy to locate the cordinates using google geocode api

from geopy.geocoders import GoogleV3

def get_lat_long(location, api_key):
    geolocator = GoogleV3(api_key=api_key)
    location_data = geolocator.geocode(location)
    
    if location_data is not None:
        latitude = location_data.latitude
        longitude = location_data.longitude
        return latitude, longitude
    else:
        print(f"Couldn't find coordinates for {location}")
        return None, None

# Read API key from secrets.txt
with open('secrets.txt') as file:
    api_key = file.read().split('=')[1].strip()

# Assuming locations is your DataFrame containing the unique locations in the 'new_location_name' column
for index, row in locations.iterrows():
    location = row['new_location_name']
    latitude, longitude = get_lat_long(location, api_key)
    
    # Update the DataFrame with latitude and longitude
    locations.at[index, 'Latitude'] = latitude
    locations.at[index, 'Longitude'] = longitude

# Print the updated DataFrame
locations
```

### Handling outliers for the column cost_two person

```python
def filter_outliers_iqr(df_model, column_name):
    """
    Filter outliers from a DataFrame column using the Interquartile Range (IQR) method.

    Parameters:
    - df (DataFrame): The DataFrame containing the data.
    - column_name (str): The name of the column from which to filter outliers.

    Returns:
    - DataFrame: A new DataFrame with outliers filtered.
    """

    # Calculate the first quartile (Q1), third quartile (Q3), and IQR for the specified column
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter the DataFrame to remove outliers
    # Keep only the rows where the column value is within the defined bounds
    filtered_df = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]

    return filtered_df
```

### Feature Engineering

```python
# creating a function to simplify and generalize location by keeping only the most frequent values in a specified column and replacing less frequent values with a common label ('Other' by default).
```

```python
def generalize_top_n_values(df, column_name, top_values_count, new_value='Other'):
    """
    Generalize a DataFrame column by keeping the top N most frequent values and replacing the rest with a new value.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the column to be generalized.
        column_name (str): The name of the column to be generalized.
        top_values_count (int): The number of top values to keep.
        new_value (str, optional): The value to replace less frequent values with (default is 'Other').

    Returns:
        pd.DataFrame: The DataFrame with the column generalized.
    """
    # Count the occurrences of each unique value in the specified column
    value_counts = df[column_name].value_counts()

    # Extract the top N values based on their frequency
    top_values = value_counts.index[:top_values_count]

    # Replace values in the column with 'new_value' for those not in the top N
    df[column_name] = df[column_name].apply(lambda x: x if x in top_values else new_value)

    return df
```

### Working on the Target Feature

Column 'rest_rating' is our target feature because it is the indicator of customer satisfaction and how well the restaurant is doing.

```python

# The mask is created based on the 'rest_rating' column in your dataset.
# The new column 'success' is set to 1 where the rating is equal to or greater than 3.75, and 0 otherwise.
# The 'success' column is then converted to an integer type for consistency and to facilitate its use in machine learning models.
```

```python
# Create a mask to identify rows where 'rest_rating' is greater than or equal to 3.75
mask = df_encoded['rest_rating'] >= 3.75
# Set the 'success' column to 1 for rows where 'rest_rating' is >= 3.75
df_encoded.loc[mask, 'success'] = 1
# Set the 'success' column to 0 for rows where 'rest_rating' is < 3.75
df_encoded.loc[~mask, 'success'] = 0
# Convert the 'success' column to integer type (specifically, int32)
df_encoded['success'] = df_encoded['success'].astype('int32')
```

```python
def plot_roc_curves(classifier_dict, X_test, y_test):
    """
    Plots ROC curves for multiple classifiers in one plot.

    Parameters:
    classifier_dict (dict): A dictionary where keys are classifier names and values are fitted classifier models with best parameters.
    X_test (array-like): Test feature data.
    y_test (array-like): True labels for the test data.
    """

    plt.figure(figsize=(8, 6))
    plt.plot([0, 1], [0, 1], 'k--', lw=2)  # Diagonal line
    
    for clf_name, clf in classifier_dict.items():
        fpr, tpr, _ = roc_curve(y_test, clf.predict_proba(X_test)[:, 1])
        roc_auc = roc_auc_score(y_test, clf.predict_proba(X_test)[:, 1])
        plt.plot(fpr, tpr, label=f'{clf_name} (AUC = {roc_auc:.2f})')

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves')
    plt.legend(loc="lower right")
    plt.show()
# Example usage:
# classifier_dict = {'Classifier1': clf1, 'Classifier2': clf2, 'Classifier3': clf3}
# plot_roc_curves(classifier_dict, X_test, y_test)
```

```python
# splitting data for training and testing

x_train, x_test, y_train, y_test = train_test_split(XScaled, y, test_size=0.3, random_state=42,shuffle=True,stratify=y)
print('x_train size:',x_train.shape)
print('y_train size:',y_train.shape)
print('x_test size:',x_test.shape)
print('x_test size:',y_test.shape)
```

```python
# Now, use value_counts to get the distribution of classes
class_distribution = y_train.value_counts()

# To get the distribution in percentage, you can use normalize=True
percentage_distribution = y_train.value_counts(normalize=True)

print("Class Distribution:\n", class_distribution)
print("\nPercentage Distribution:\n", percentage_distribution * 100)
```

```python
## Upsampling the train set using SMOTE to increase accuracy of the model

from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(x_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV,train_test_split,RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
from sklearn.metrics import classification_report,confusion_matrix,recall_score,precision_score,f1_score,roc_curve,accuracy_score
from sklearn.metrics import roc_auc_score,precision_recall_curve,ConfusionMatrixDisplay,make_scorer
```

```python
# defining your machine learning algorithms
algorithms = {
    'Random Forest': {
        'model': RandomForestClassifier(),
        'param_grid': {
                'bootstrap': [True, False],
                'max_depth': [3, 5, 10, 20, 50],
                'n_estimators': [50, 100, 200, 500],
                'max_features': ['auto', 'sqrt']
        }
    },
    'K-Nearest Neighbors': {
        'model': KNeighborsClassifier(),
        'param_grid': {
                'n_neighbors': [7, 9, 11, 13],
                'weights': ['uniform', 'distance']
        }
    },
    'Logistic Regression': {
        'model': LogisticRegression(),
        'param_grid': {
                'C': np.linspace(0.1, 10, 20),
                'penalty': ['l1', 'l2'],
                'solver': ['liblinear']
        }
    },
     'Decision Trees': {
        'model': DecisionTreeClassifier(),
        'param_grid': {
                'criterion': ['gini', 'entropy'],
                'max_depth': [10, 20, 30, 50],  
                'min_samples_split': [5, 10,20],
                'min_samples_leaf': [2, 4,6]
        }
    }
}
```

```python
best_models = {}
# Create an F1 scorer for use in GridSearchCV
scorer = make_scorer(f1_score)

# Threshold for lowering to increase recall
threshold = 0.5

# Function to evaluate metrics - make sure you define this before calling it
def Evaluation_Metrices(y_true, y_pred, dataset_name):
    print(f"***** Evaluation Metrics for {dataset_name} Data *****")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall:", recall_score(y_true, y_pred))
    print("F1 Score:", f1_score(y_true, y_pred))
    print("************************************************")

# Iterate through each algorithm and perform GridSearchCV
for algo_name, algo_info in algorithms.items():
    model = algo_info['model']
    param_grid = algo_info['param_grid']
    
    print(f'*********** Fitting {algo_name} ***********')
    
    # Initialize GridSearchCV for hyperparameter tuning
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring=scorer, n_jobs=-1, verbose=3)
    # Fit the model to the training data
    grid_search.fit(X_resampled, y_resampled)
    
    # Save the best model for this algorithm
    best_models[algo_name] = grid_search.best_estimator_
    
    # Evaluate the best model's performance on the test set
    y_prob = best_models[algo_name].predict_proba(x_test)[:, 1]  # Get predicted probabilities for the positive class
    y_pred = (y_prob > threshold).astype(int)  # Apply thresholding to predict based on probability
    
    # Call the evaluation function
    Evaluation_Metrices(y_test, y_pred, algo_name)
    
    print(f'*******************************************\n')
```

```python
plot_roc_curves(best_models, x_test, y_test)
```

Random Forest (AUC = 0.82): The Random Forest model has the highest AUC of all four models, indicating that it is the most effective at distinguishing between the positive and negative classes. Its curve is closer to the top-left corner of the plot, which signifies a higher true positive rate and a lower false positive rate.

The Random Forest algorithm emerged as the best viable option with accuracy of 82%. Furthermore, the balance between precision and recall was analyzed and considered to be within acceptable limits for the project's objectives.
